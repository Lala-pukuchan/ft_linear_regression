```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('../input/data.csv')

# get all the data from the first column
X = data.iloc[:, 0]
# get all the data from the second column
Y = data.iloc[:, 1]

# Plot the data
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
# initial value
theta1 = 0
theta0 = 0

# learning rate
L = 0.0001
epochs = 1000

# number of data
n = float(len(X))

# List to store mean squared errors
mse_values = []

for i in range(epochs):
    Y_pred = theta1 * X + theta0
    D_theta1 = (-2/n) * sum(X * (Y - Y_pred))
    D_theta0 = (-2/n) * sum(Y - Y_pred)
    theta1 = theta1 - L * D_theta1
    theta0 = theta0 - L*D_theta0

print(theta1, theta0)
```

```python
# Making predections
Y_pred = theta1*X + theta0

# Convert to Numpy arrays
X_np = X.to_numpy()
Y_pred_np = Y_pred.to_numpy()

# Plot the data and the regression line
plt.scatter(X, Y)
plt.plot(X_np, Y_pred_np, color='red')
plt.show()
```
